```python
# function to load weight metrics
def load_metrics(xml_path):
    tree = ET.parse(xml_path)
    num_node = len(root.findall(".//vertex"))
    weights_metric = np.nan_to_num(np.identity(num_node) * 0, 0.0)

    for i, vertex in enumerate(root.findall('.//vertex')):
        for edge in vertex.findall('.//edge'):
            cost = float(edge.get("cost"))
            node = int(edge.text)
            weights_metric[i,node] = cost


    return weights_metric
```

Prepare function for EA

```python
# Calculate the fitness of each individual
def calculate_total_distance(solution, distance_matrix):
    total_distance = 0
    num_cities = len(solution)
    for i in range(num_cities):
        from_city = solution[i]
        to_city = solution[(i + 1) % num_cities]  # Wrap around to the first city
        total_distance += distance_matrix[from_city][to_city]
    return total_distance
```

```python
def tournament_selection(population, fitness_scores, tournament_size):
    # Perform tournament selection to choose one individual
    tournament = random.sample(range(len(population)), tournament_size)
    best_individual = min(tournament, key=lambda i: fitness_scores[i])
    return population[best_individual]
```

```python
def single_point_crossover(parent_a: List[str], parent_b: List[str], locus: Optional[int] = None):
    n = len(parent_a)
    if locus is None:
        locus = np.random.randint(1, n)
    
    child_c = []
    child_d = []
    
    for i in range(locus):
        if parent_a[i] not in child_c:
            child_c.append(parent_a[i])
        if parent_b[i] not in child_d:
            child_d.append(parent_b[i])
    
    for i in range(n):
        if parent_b[i] not in child_c:
            child_c.append(parent_b[i])
        if parent_a[i] not in child_d:
            child_d.append(parent_a[i])
    
    return child_c, child_d,
```

```python
def mutate(solution, mutation_rate):
    mutated_solution = solution.copy()  # Create a copy of the solution

    if random.random() < mutation_rate:
        # Select two random distinct positions in the solution
        pos1, pos2 = random.sample(range(len(mutated_solution)), 2)

        # Swap the cities at pos1 and pos2
        mutated_solution[pos1], mutated_solution[pos2] = mutated_solution[pos2], mutated_solution[pos1]

    return mutated_solution
```

```python
# Define a replacement function that returns the updated population
def replacement(population, fitness_scores, new_individual, new_fitness):
    # Find the index of the worst individual in the current population
    worst_index = fitness_scores.index(max(fitness_scores))

    # Check if the new individual's fitness is better than the worst individual in the population
    if new_fitness < fitness_scores[worst_index]:
        # Replace the worst individual with the new individual
        population[worst_index] = new_individual
        fitness_scores[worst_index]= new_fitness

    return population , fitness_scores
```

```python
def run_evolutionary_algorithm(num_iterations, population, tournament_size, mutation_rate, distance_matrix):
    Min_Fitness = []
    Avg_Fitness = []
    fitness_scores = [calculate_total_distance(individual, distance_matrix) for individual in population]
    

   

    for iteration in range(num_iterations):
        

        parent_a = tournament_selection(population, fitness_scores, tournament_size)
        parent_b = tournament_selection(population, fitness_scores, tournament_size)

        child_c, child_d = single_point_crossover(parent_a, parent_b)

        e = mutate(child_c, mutation_rate)
        f = mutate(child_d, mutation_rate)

        fitness_e = calculate_total_distance(e, distance_matrix)
        fitness_f = calculate_total_distance(f, distance_matrix)

        replacement(population, fitness_scores, e, fitness_e)
        replacement(population, fitness_scores, f, fitness_f)

        fitness_scores = [calculate_total_distance(individual, distance_matrix) for individual in population]

        min_fitness = min(fitness_scores)
        avg_fitness = sum(fitness_scores) / len(fitness_scores)
        Min_Fitness.append(min_fitness)
        Avg_Fitness.append(avg_fitness)
        



    return Min_Fitness, Avg_Fitness
```

Set parameter

```python
num_cities_burma = len(burma_metrics)
num_cities_brazil = len(brazil_metrics)
# Population size
p = 100
tournament_size = 5
mutation_rate = 0.2

population_burma = []
population_brazil = []
```

```python
for _ in range(p):
    # Generate a random permutation of city indices
    random_solution = random.sample(range(num_cities_burma), num_cities_burma)
    population_burma.append(random_solution)
```

```python
for _ in range(p):
    # Generate a random permutation of city indices
    random_solution = random.sample(range(num_cities_brazil), num_cities_brazil)
    population_brazil.append(random_solution)
```

loop operation

```python
num_iterations = 10000
```

```python
Min_Fitness_burma, Avg_Fitness_burma = run_evolutionary_algorithm(num_iterations, population_burma, tournament_size, mutation_rate, burma_metrics)
```

```python
Min_Fitness_brazil, Avg_Fitness_brazil= run_evolutionary_algorithm(num_iterations, population_brazil, tournament_size, mutation_rate, brazil_metrics)
```

```python
def index_of_first_lowest_value(values):
    # Return the index of the first occurrence of the minimum value
    return values.index(min(values))
```

```python
Convergence_burma=index_of_first_lowest_value(Min_Fitness_burma)
Convergence_brazil=index_of_first_lowest_value(Min_Fitness_brazil)
```

```python
# Create a list of iteration numbers for the x-axis
iterations = list(range(0, num_iterations + 0))

# Create two subplots side by side
plt.figure(figsize=(12, 6))

# Subplot 1 for Burma
plt.subplot(1, 2, 1)
plt.plot(iterations, Min_Fitness_burma, label="Min Fitness (Burma)")
plt.plot(iterations, Avg_Fitness_burma, label="Avg Fitness (Burma)")
plt.axvline(x=Convergence_burma, color='r', linestyle='--', label="Convergence Line")
plt.xlabel("Iteration")
plt.xlim(0, 1000)
plt.ylabel("Fitness")
plt.legend()
plt.title("Min and Avg Fitness Over Iterations (Burma)")
plt.grid(True)

# Subplot 2 for Brazil
plt.subplot(1, 2, 2)
plt.plot(iterations, Min_Fitness_brazil, label="Min Fitness (Brazil)")
plt.plot(iterations, Avg_Fitness_brazil, label="Avg Fitness (Brazil)")
plt.axvline(x=Convergence_brazil, color='r', linestyle='--', label="Convergence Line")
plt.xlabel("Iteration")
plt.ylabel("Fitness")
plt.legend()
plt.title("Min and Avg Fitness Over Iterations (Brazil)")
plt.grid(True)

# Adjust the layout for better spacing between the two subplots
plt.tight_layout()

plt.show()
```
