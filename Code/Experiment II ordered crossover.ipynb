```python
# function to load weight metrics
def load_metrics(xml_path):
    tree = ET.parse(xml_path)
    num_node = len(root.findall(".//vertex"))
    weights_metric = np.nan_to_num(np.identity(num_node) * 0, 0.0)

    for i, vertex in enumerate(root.findall('.//vertex')):
        for edge in vertex.findall('.//edge'):
            cost = float(edge.get("cost"))
            node = int(edge.text)
            weights_metric[i,node] = cost


    return weights_metric
```

Prepare function for EA

```python
# Calculate the fitness of each individual
def calculate_total_distance(solution, distance_matrix):
    total_distance = 0
    num_cities = len(solution)
    for i in range(num_cities):
        from_city = solution[i]
        to_city = solution[(i + 1) % num_cities]  # Wrap around to the first city
        total_distance += distance_matrix[from_city][to_city]
    return total_distance
```

```python
def tournament_selection(population, fitness_scores, tournament_size):
    # Perform tournament selection to choose one individual
    tournament = random.sample(range(len(population)), tournament_size)
    best_individual = min(tournament, key=lambda i: fitness_scores[i])
    return population[best_individual]
```

```python
def tsp_two_point_crossover(parent_a: List[str], parent_b: List[str], locus1: Optional[int] = None, locus2: Optional[int] = None) -> (List[str], List[str]):
    n = len(parent_a)
    if locus1 is None or locus2 is None:
        locus1, locus2 = sorted(np.random.choice(range(1, n), 2, replace=False))
    
    # Initialize children with None
    child_c = [None] * n
    child_d = [None] * n

    # Copy the segment between crossover points from one parent to one child
    child_c[locus1:locus2] = parent_a[locus1:locus2]
    child_d[locus1:locus2] = parent_b[locus1:locus2]

    # Function to fill remaining cities
    def fill_remaining_cities(child, other_parent):
        current_pos = locus2
        for city in other_parent:
            # Skip cities already in the child
            if city not in child:
                # Place city in the next available position
                while child[current_pos % n] is not None:
                    current_pos += 1
                child[current_pos % n] = city

    fill_remaining_cities(child_c, parent_b)
    fill_remaining_cities(child_d, parent_a)
    
    return child_c, child_d
```

```python
def mutate(solution, mutation_rate):
    mutated_solution = solution.copy()  # Create a copy of the solution

    if random.random() < mutation_rate:
        # Select two random distinct positions in the solution
        pos1, pos2 = random.sample(range(len(mutated_solution)), 2)

        # Swap the cities at pos1 and pos2
        mutated_solution[pos1], mutated_solution[pos2] = mutated_solution[pos2], mutated_solution[pos1]

    return mutated_solution
```

```python
# Define a replacement function that returns the updated population
def replacement(population, fitness_scores, new_individual, new_fitness):
    # Find the index of the worst individual in the current population
    worst_index = fitness_scores.index(max(fitness_scores))

    # Check if the new individual's fitness is better than the worst individual in the population
    if new_fitness < fitness_scores[worst_index]:
        # Replace the worst individual with the new individual
        population[worst_index] = new_individual
        fitness_scores[worst_index]= new_fitness

    return population , fitness_scores
```

```python

def generate_initial_population(population_size, distance_matrix):
    num_cities = len(distance_matrix)
    population = []

    for _ in range(population_size):
        individual = list(range(num_cities))
        random.shuffle(individual)
        population.append(individual)

    return population
```

```python
def index_of_first_lowest_value(values):
    # Return the index of the first occurrence of the minimum value
    return values.index(min(values))
```

```python
def run_evolutionary_algorithm(num_iterations, population, tournament_size, mutation_rate, distance_matrix):
    Min_Fitness = []
    Avg_Fitness = []
    fitness_scores = [calculate_total_distance(individual, distance_matrix) for individual in population]
   

    for iteration in range(num_iterations):
        

        parent_a = tournament_selection(population, fitness_scores, tournament_size)
        parent_b = tournament_selection(population, fitness_scores, tournament_size)

        child_c, child_d =  tsp_two_point_crossover(parent_a, parent_b)

        e = mutate(child_c, mutation_rate)
        f = mutate(child_d, mutation_rate)

        fitness_e = calculate_total_distance(e, distance_matrix)
        fitness_f = calculate_total_distance(f, distance_matrix)

        replacement(population, fitness_scores, e, fitness_e)
        replacement(population, fitness_scores, f, fitness_f)

        fitness_scores = [calculate_total_distance(individual, distance_matrix) for individual in population]

        min_fitness = min(fitness_scores)
        avg_fitness = sum(fitness_scores) / len(fitness_scores)
        Min_Fitness.append(min_fitness)
    Convergence_point=index_of_first_lowest_value(Min_Fitness)

    return min_fitness, avg_fitness, Convergence_point
```

```python
def run_multiple_trials(num_trials, num_iterations, population_size, tournament_size, mutation_rate, distance_matrix):
    # Start timing
    start_time = time.time()
    
    all_min_fitness = []
    all_avg_fitness = []
    all_convergence_point = []


    for trial in range(num_trials):
        # Initialize or generate a new population for each trial
        population = generate_initial_population(population_size, distance_matrix)

        # Run the evolutionary algorithm
        min_fitness, avg_fitness, Convergence_point = run_evolutionary_algorithm(num_iterations, population, tournament_size, mutation_rate, distance_matrix)

        # Store the results
        all_min_fitness.append(min_fitness)
        all_avg_fitness.append(avg_fitness)
        all_convergence_point.append(Convergence_point)

        # Print out the results of each trial
        print(f"Trial {trial + 1}: Min Fitness = {min_fitness}, Avg Fitness = {avg_fitness}, Convergence Point = {Convergence_point}")

    # Calculate overall statistics
    overall_min_fitness_avg = sum(all_min_fitness) / num_trials
    overall_avg_fitness_avg = sum(all_avg_fitness) / num_trials
    
        # End timing
    end_time = time.time()
    runtime = end_time - start_time

    print(f"Total Runtime: {runtime} seconds")

    return all_min_fitness, all_avg_fitness, all_convergence_point
```

Set parameter

```python
# burma
num_trials = 20
num_iterations = 10000
population_size = 50
tournament_size = 5
mutation_rate = 0.1
distance_matrix = burma_metrics


all_min_fitness_burma, all_avg_fitness_burma, all_CP_burma = run_multiple_trials(num_trials, num_iterations, population_size, tournament_size, mutation_rate, distance_matrix)


all_min_fitness_burma = np.array(all_min_fitness_burma)
all_avg_fitness_burma = np.array(all_avg_fitness_burma)
all_CP_burma = np.array(all_CP_burma)
overall_min=np.average(all_min_fitness_burma)
overall_avg=np.average(all_avg_fitness_burma)
overall_CP=np.average(all_CP_burma)
print(f"Average of Min Fitness over {num_trials} trials: {overall_min}")
print(f"Average of Avg Fitness over {num_trials} trials: {overall_avg}")
print(f"Average of Convergence point over {num_trials} trials: {overall_CP}")
```
